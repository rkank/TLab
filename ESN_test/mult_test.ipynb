```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm, trange
```

```python
squared_losses = []
losses = []

for iter in trange(1000):
	loss = 0
	squared_loss = 0
	count = 0
	for i in range(1, 101):
		for j in range(1, 101):
			loss += ((np.random.randint(1, 100) * np.random.randint(1, 100)) - (i * j))
			squared_loss += ((np.random.randint(1, 100) * np.random.randint(1, 100)) - (i * j))**2
			count += 1

	squared_losses.append(squared_loss / count)
	losses.append(loss / count)
```

```python
print(sum(squared_losses)/len(squared_losses))
print(sum(losses)/len(losses))
```

```python
plt.plot(losses)
```

```python
plt.semilogy(squared_losses)
```
